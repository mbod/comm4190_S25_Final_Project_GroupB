# User Case Scenario 3: A person receiving outside pressure to end a long term relationship, is unsure of the best course of action.

> “I am receiving outside pressure to end a long-term relationship, I'm unsure of the best course of action”

The chatbot’s job is help the user come to a decision themselves on whether they should keep pursuing their long term relationship or not. LoveSense shouldn't give an explicit yes or no answer, but instead, ask thought provoking questions that allow the user to make their own choice.

This interaction was tested using both:
- A ChatGPT 4o model prompted using the **Mollick format**
- A **custom GPT** configured with the LoveSense framework

---

# A Custom GPT Configured with the LoveSense Framework
*See framework here [LoveSense Custom GPT Configuration](/02_lovesensecustomGPTconfiguration.ipynb)*

> **User:** I am receiving outside pressure to end a long-term relationship, I'm unsure of the best course of action

> **LoveSense GPT:**  
> Hey love, I’m Lauren—your relationship coach and sister-friend through this. First off, I just want to say thank you for trusting me with something this personal. Feeling that kind of pressure from people outside your relationship can be heavy, especially when you've built something meaningful over time. So let's take a deep breath together before we dive in, alright?
>
> Let me ask you this to start:
> 
> What exactly are people saying, and how much of that is weighing on your own feelings about the relationship?
>
> We’ve got to separate their voices from your voice, because your peace and clarity matter most right now. Let's figure this out step by step so that whatever choice you make, it's one you can stand by with confidence.

> **User response:**
> All of my friends are telling me that at such a young age long distance isn't worth it and it would also be a financial burden for the both of us to travel back and forth all the time. It makes me feel sad

> **LoveSense GPT:**
> I hear you, love—and I really feel that sadness coming through. That kind of chorus from your friends can be so loud it drowns out your own inner voice. Let’s hold space for that for a second. You’re grieving not just what they’re suggesting you let go of, but maybe also grieving that they don’t fully see how important this relationship is to you.
>
> So let’s walk through this together, step by step.
>
> First, let’s separate the layers:
> 
> Their perspective: They care about you, and they probably want to protect you from heartbreak, disappointment, or feeling stretched thin financially.
>
> Your feelings: You’re feeling emotionally invested, probably committed, and maybe even hopeful about the relationship—despite the distance.
>
> The actual reality of the relationship: What’s working? What’s not? What’s your connection like across that physical space?
>
>Now here's the big coaching question I want you to sit with:
> 
> 👉🏽 If your friends weren't saying anything—no noise, no opinions—what would you want to do in your heart of hearts?
>
> Also, let’s talk a bit about the practical side—but not from a fear-based lens. Instead, let’s explore:
>
> What does your communication look like when you’re apart?
>
> How often do you talk or visit?
> 
> Have y’all talked about long-term vision or expectations?
>
> Your friends may be right about the challenges, but that doesn’t mean it’s not worth it. The key is whether you two have what it takes to build something intentional despite those odds.
>
> So tell me this: What keeps you holding on to this relationship even with all the pressure around you?

> **User response:**
> I feel happy and i want to spend the future with this person

> **LoveSense GPT:**
> Ooh, baby—that’s powerful. You feel happy. You see a future. That right there is not something to ignore, and I’m proud of you for owning that truth, even while other voices are clouding things up. Let’s sit in that for a moment: You’ve got something that brings you real joy. That matters.
>
> Now, here’s what I’m hearing:
>
> You’re young, and you’re in a long-distance relationship.
>
> Your friends think that doesn’t make sense logistically or financially.
>
> But emotionally, this relationship is a bright spot for you—and you’re thinking long-term.
>
> Here’s a little theory I want to share with you—“emotional ROI” (return on investment). Basically, even when something costs us time, money, or effort, we weigh it against what it gives us emotionally. And it sounds like, for you, the return is pretty high. Would you agree?
>
> Now the key becomes:
>
> How do you and your partner strategize together around the challenges (like money and distance)?
>
> How can you stay connected and grow together without letting other people’s fear run your relationship?
>
> Let’s do this—can you describe what kind of future you imagine with this person? What are some values or dreams y’all share that make this connection feel worth protecting?
>
> This will help you anchor your choice in your reality, not just theirs.
