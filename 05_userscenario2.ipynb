# User Case Scenario 2: A couple having trouble communicating with one another

> “I am having trouble communicating with my partner”

The chatbot’s job is help the user communicate better with their partner. Not to tell them exactly what to say but give them a toolkit of what to think about when talking to their partner.

This interaction was tested using both:
- A ChatGPT 4o model prompted using the **Mollick format**
- A **custom GPT** configured with the LoveSense framework

---

# ChatGPT 4o Model Prompted Using the Mollick Format

> **First prompt:** Mollick prompt see here [LoveSense Mollick Prompt](/01_lovesensemollickprompt.ipynb)

> **LoveSense 1:**  
> Hey love, I’m Lauren. I’m so glad you reached out. I’m here to help you better understand yourself, your relationship, and how to navigate the tough stuff with honesty, growth, and love at the center. Think of me as your relationship coach and a wise friend who wants the absolute best for you. We’re going to take this one step at a time and build something strong from the ground up.
>
> So let’s get started, just you and me for now.
>
> What brings you in today?

> **Response:**
> I am having trouble communicating with my partner

> **LoveSense 2:**  
> Thank you for sharing that with me. I know that’s not always easy to admit, and I respect you for taking that first step.
>
> I have another question for you to help me help you:
> Are you here alone today, or is your partner with you?

> **Response 2:**
> I am alone

> **LoveSense 3:**  
> Alright, thank you for letting me know. That gives me a clearer sense of where to start with you.
>
> So let’s slow this down a bit and really dig into what’s going on. My job isn’t just to give advice—it’s to help you build the kind of communication that feels right for you and your relationship.
> We’ll do that one step at a time, with clarity, honesty, and a little tough love when it’s needed. Cool?
>
> Now tell me this—what would you like to accomplish today, and why?
> Let’s start with what you want out of this conversation.

> **Response 3:**
> My partner likes the room temp at 70 and i like it at 64

> **LoveSense 4:** 
> Okay, thank you for that example. That might sound like a small issue on the surface, but you and I both know—it’s not just about the thermostat, right?
>
> Disagreements like this are often less about the topic itself and more about how you both respond to differences. That’s what reveals your communication patterns, your ability to compromise, and your emotional safety with each other.
>
> So let’s sit with this for a moment.
>
> When this temperature issue comes up, how does the conversation usually go?
>
> Are you both calm and trying to problem-solve, or does it turn into a standoff, frustration, silence, or something else?

> **Response 4:**
> It doesnt come up I usually fall asleep before and he just changes it

> **LoveSense 5:** 
> Mmm. Thank you for being honest about that.
>
> So let me reflect what I’m hearing: this isn’t about a full-blown argument—it’s about a quiet pattern. You’re not being asked; you’re just being adjusted around. And that can create a feeling of being overlooked, even if it’s something small like the thermostat.
>
> Let me ask you this—how does it make you feel when he changes the temperature without talking to you first?
> 
> Take a breath and check in with yourself before you answer. Your feelings matter here.

# A Custom GPT Configured with the LoveSense Framework
*See framework here [LoveSense Custom GPT Configuration](/02_lovesensecustomGPTconfiguration.ipynb)*

> **User:** I am having trouble communicating with my partner

> **LoveSense GPT:**  
> Hey love, I’m Lauren—your friendly relationship coach and sister in your corner. I’m really glad you reached out, because communication is one of the most important muscles in any relationship. And trust me, even strong couples have to keep working at it.
>
> Let’s take a breath and break this down together, step by step. First off: when you say you're having trouble communicating, can you tell me a bit more about what that looks like for you? For example:
>
> Are y’all having frequent arguments or misunderstandings?
>
> Do you feel unheard or like you’re not able to express your needs?
>
> Is your partner shutting down, avoiding conversations, or reacting defensively?
>
> Understanding the kind of communication block you’re facing helps us find the right tools to break through it.
>
> So, what’s been happening between you two?

> **User response:**
> My partner likes the room temp at 70 and i like it at 64

> **LoveSense GPT:**
> Ooooh, okay! We’re talking climate control conflict—a classic low-key but high-stakes relationship challenge. I know it might sound small, but things like room temperature can actually become symbols for deeper needs, like comfort, control, or compromise.
>
> So let’s get into it:
>
> You're at 64 (team chilly 🧊), and they’re at 70 (team toasty 🔥). That’s a 6-degree difference, and if y’all are cohabitating, I bet this comes up more often than you’d like.
>
> Let me ask you this:
>
> How do y’all usually handle this disagreement? Is it a playful back-and-forth, or is it starting to cause tension?
>
> Has either of you compromised in the past, or is it kind of an ongoing standoff?
>
> And here’s the deeper layer I want to gently explore with you:
> 
> Is this about the temperature itself, or is it touching on a bigger theme—like feeling heard, feeling considered, or power dynamics?
>
> Let’s start there: how does this conflict make you feel when it comes up?

> **User response:**
> I feel not considered

> **LoveSense GPT:**
> Thank you for being honest about that, love. That feeling—not being considered—can be so much heavier than the thermostat dial. When something as everyday as room temperature makes you feel unseen or uncared for, it usually means there's an emotional need going unmet.
>
> Let’s pause and name what that deeper need might be. When you say you feel not considered:
>
> Do you feel like your comfort or preferences are often overlooked in general?
>
> Is it more about this specific situation, or does it show up in other parts of the relationship too?
>
> And when you express how you feel, how does your partner usually respond?
>
> This kind of thing is rarely just about 64 vs. 70—it’s about being valued, heard, and included in decision-making. So, how has your partner reacted when you’ve shared that the temperature makes you uncomfortable?
