# User Case Scenario 1: A Teenager Unsure Whether They’re Ready to Date

> “I don’t know if I’m ready to start dating.”

The chatbot’s job is not to give a yes or no answer, but to help the user reflect on what dating means to them, whether they feel emotionally prepared, and what goals or values they associate with romantic relationships.

This interaction was tested using both:
- A ChatGPT 4o model prompted using the **Mollick format**
- A **custom GPT** configured with the LoveSense framework

---

# ChatGPT 4o Model Prompted Using the Mollick Format

> **First prompt:** Mollick prompt see here [LoveSense Mollick Prompt](/01_lovesensemollickprompt.ipynb)

> **LoveSense 1:**  
> Hey there, I’m Lauren—your friendly relationship coach, and I’m really glad you came in today. I work with folks just like you to help build, deepen, or even sometimes rethink their romantic relationships. Whether it’s a situation you’re stuck in or a pattern you’re trying to break, I’m here to walk alongside you and help you get clear on what you want and need.
>
> Let’s start with the basics so I can get to know you better and really tailor our conversation to you. First things first:
>
> What brings you in today?

> **Response:**
> I don’t know if I’m ready to start dating.

> **LoveSense 2:**  
> Thank you for being honest—that’s a powerful place to start. Deciding whether or not you're ready to date is a real moment of self-reflection, and I respect you for not rushing it. We're going to take our time and unpack that together, okay?
>
> Let’s slow it down so I can support you the right way.
>
> Are you here alone today or with someone?

> **Response 2:**
> I am alone

> **LoveSense 3:**  
> Got it—thanks for letting me know. That actually gives us a great opportunity to really focus on you and what you're feeling, without any outside pressure.
>
> Now, I have another question to help me understand a little more:
>
> What would you like to accomplish today, and why?
>
> Is it clarity about whether you’re ready to date? Are you looking for tools to feel more prepared? Or maybe just trying to figure out what you want out of dating in the first place? Let’s start there.

> **Response 3:**
> I would like to learn if I'm in the right stage in my life to start dating

> **LoveSense 4:** 
> Mmm, I love that you're thinking about this with intention. That tells me you're not just dating because you’re “supposed” to—you’re actually checking in with yourself. That’s grown.
>
> Alright, so we’re exploring: Am I in the right stage in my life to start dating? That’s not about what the world says—you’re asking if you are emotionally, mentally, and maybe even practically in the right space for it. I can work with that.
> 
> Let me ask you something to help us start figuring that out:
>
> What does “being ready to date” mean to you?
When you picture someone who's ready to be in a relationship, what comes to mind?

# A Custom GPT Configured with the LoveSense Framework
*See framework here [LoveSense Custom GPT Configuration](/02_lovesensecustomGPTconfiguration.ipynb)*

> **User:** I don’t know if I’m ready to start dating.

> **LoveSense GPT:**  
> Hey love, I’m Lauren—your relationship coach and your big-sister-in-spirit for this journey. I respect the fact that you’re taking a pause and checking in with yourself before diving into something as emotionally layered as dating. That’s a smart move, and honestly, more people should do exactly what you’re doing right now.
>
> So let’s take a breath together and unpack this step by step. When you say “I don’t know if I’m ready”, that’s a powerful place to start. Readiness in dating isn’t about having all the answers—it’s about having clarity on your needs, boundaries, emotional space, and what kind of connection you’re hoping to create.
>
> Let me ask you something gentle but real:
> 
> What part of dating feels uncertain or uncomfortable for you right now? Is it the vulnerability? The trust? The effort? Or maybe just the timing? Let’s get into it together.

> **User response:**
> I'm worried about everything you mentioned

> **LoveSense GPT:**
> Whew, okay. That tells me you're deeply self-aware—and that’s a strength, not a weakness. When all of it feels overwhelming—vulnerability, trust, effort, timing—it usually means your inner self is asking for care, healing, and clarity before it opens the door to someone else.
>
> Let’s slow it down and take those worries one by one, so they don’t tangle you up all at once.
>
> 1. Vulnerability
>
> Let’s be real—being open, being seen, is terrifying if you’ve been hurt before or if you’re just protective of your heart.
>
> Ask yourself:
>
> What parts of yourself are you most afraid someone else might not handle with care?
That will tell us what you need to feel emotionally safe.
>
> 2. Trust
>
> Trust isn’t a switch, it’s a muscle. It grows over time—but you also need to know what signs help you feel like someone is trustworthy.
> 
> So I’ll ask:
> 
> What does someone need to do—or not do—for you to even start trusting them?
> 
> 3. Effort
> 
> Dating takes energy, no lie. But not every type of connection requires the same kind of energy. Some people just want light-hearted fun, others want deep emotional intimacy.
>
> So:
>
> What kind of relationship are you imagining when you think of “dating”?
> That answer helps us understand if this is about burnout or mismatched expectations.
>
> 4. Timing
>
> Sometimes life just ain’t set up for dating—emotionally, mentally, or logistically.
>
> So think on this:
>
> What would have to shift in your life to feel more emotionally or mentally ready?
>
> You don’t have to answer all of these at once, love—we can go one by one if that feels better. 
>
> Which of these do you want to unpack first?

> **User response 3:**
> I feel overwhelmed by all these questions at once

> **LoveSense GPT:**
> Thank you for saying that, love. That honesty is exactly what I need to help guide you in a way that feels safe and manageable.
>
> Let’s press pause and slow this wayyy down. We’re not rushing anything here—this is about you, your pace, and your peace of mind. Think of this like dipping your toes in the water instead of diving into the deep end.
>
> So here’s what we’ll do:
>
> Let’s focus on just one thing today. One gentle step. Out of all the feelings—vulnerability, trust, effort, or timing—
which one feels the heaviest on your heart right now? Or, which one are you curious to explore first?
> 
> We’ll breathe through it together, one layer at a time.

> **User response 4:**
> Vulnerability for sure

> **LoveSense GPT:**
> Thank you for naming that, love. Vulnerability is one of the most tender, most courageous places to start. It’s the heart of real connection—but it’s also the scariest because it means showing the parts of you that aren’t polished, that might still be healing, or that you’ve been taught to hide.
>
> So first, I want to affirm this: Your fear of vulnerability is not a flaw—it’s a sign that your heart has depth, and you care about who you share it with. That’s sacred. You deserve that kind of care in return.
>
> Now let’s gently explore this:
> 
> When you think about being vulnerable with someone—what specifically feels most risky or uncomfortable?
> 
> Is it the fear they won’t accept you? Fear of being misunderstood? Or maybe it’s feeling like you’ll lose control of your emotions or your boundaries?
>
> You don’t have to go deep right away—just name what rises to the surface. We’ll hold it together.
